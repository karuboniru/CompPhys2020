```python
from monte_carlo import monte_carlo
import numpy as np

def getpressure(density, temp=0.9):
    # temp = 0.9
    count = 200
    step = 80000
    sim = monte_carlo(maxstep=step, temp=temp, count=count, size=np.sqrt(count/density), mode='periodic',
                  dimension=2, rand=True, pressure_record=True)
    plist = np.array(sim.get_pressure())
    average_pressure = np.average(plist)
    fluct_pressure = np.average(plist*plist) - average_pressure**2
    return average_pressure, fluct_pressure
```

```python
densitylist = np.linspace(0, 1, 20)[1:]
pre_list=[]
flu_list=[]
for i in densitylist:
    pre, flu = getpressure(i, temp=0.9)
    pre_list.append(pre)
    flu_list.append(flu)
```

```python
pre_list1=[]
flu_list1=[]
for i in densitylist:
    pre, flu = getpressure(i, temp=2.0)
    pre_list1.append(pre)
    flu_list1.append(flu)
```

```python
from matplotlib import pyplot as plt
line1, = plt.plot(densitylist, pre_list, 'b')
line1.set_label('T = 0.9')
line2, = plt.plot(densitylist, pre_list1, 'g')
line2.set_label('T = 2.0')
plt.xlabel('density')
plt.ylabel('pressure')
plt.legend()
# plt.show()
plt.savefig('fig/pressure_200.eps')
```
